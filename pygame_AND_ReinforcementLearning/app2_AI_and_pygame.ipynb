```python
import random, pygame, sys
from pygame.locals import *
import numpy as np
```

# Make AI for Game

```python
WINDOWWIDTH = 700
WINDOWHEIGHT = 700
CELLSIZE = 100
CELLWIDTH = int(WINDOWWIDTH / CELLSIZE)
CELLHEIGHT = int(WINDOWHEIGHT / CELLSIZE)
assert WINDOWWIDTH % CELLSIZE == 0, "Window width must be a multiple of cell size."
assert WINDOWHEIGHT % CELLSIZE == 0, "Window height must be a multiple of cell size."
```

## Make Q_learning

```python
#make Q_learning for mouse and cat
states = []
for i_cat in range(CELLWIDTH):
    for j_cat in range(CELLHEIGHT):
        for i_mouse in range(CELLWIDTH):
            for j_mouse in range(CELLHEIGHT):
                states.append((i_cat,j_cat,i_mouse,j_mouse))
```

Two agents are trained to play better

```python
actions = [0,1,2,3]
Q_mouse ={}
for state in states:
    for action in actions:
        Q_mouse[state,action] = 0
# Q_mouse
Q_cat ={}
for state in states:
    for action in actions:
        Q_cat[state,action] = 0
```

```python
def get_state(observattion):
    i_cat,j_cat,i_mouse,j_mouse = observattion
    return (i_cat,j_cat,i_mouse,j_mouse)
def maxAction(Q, state, actions=[0,1,2,3]):
    values = np.array([Q[state,a] for a in actions])
    action = np.argmax(values)
    return action
```

```python
def rewards_mouse(state, hole):
    rewards = -1
    if state[0]==state[2] and state[1]==state[3]:
        rewards =100
    for i in range(len(hole_)):
        if abs(state[2]-hole_[i][0])<0.1 and abs(state[3]-hole_[i][1])<0.1:
            rewards = -100
    return rewards
```

```python
def rewards_cat(state, hole,goal):
    rewards = -1
    if state[0]==goal[0][0] and state[1]==goal[0][0]:
        rewards =100
    if state[0]==state[2] and state[1]==state[3]:
        rewards =-100
    for i in range(len(hole_)):
        if abs(state[2]-hole_[i][0])<0.1 and abs(state[3]-hole_[i][1])<0.1:
            rewards = -100
    
    return rewards
```

```python
def do_action_for_mouse(action,state):
    x = state[2]
    y = state[3]
    if action ==0:
        x+=1
        if x >= CELLWIDTH:
            x = CELLWIDTH-1
    if action ==1:
        x-=1
        if x < 0:
            x = 0
    if action ==2:
        y+=1
        if y >= CELLHEIGHT:
            y = CELLHEIGHT-1
    if action ==3:
        y-=1
        if y < 0:
            y = 0
    return (state[0],state[1],x,y)
```

```python
def do_action_for_cat(action,state):
    x = state[0]
    y = state[1]
    if action ==0:
        x+=1
        if x >= CELLWIDTH:
            x = CELLWIDTH-1
    if action ==1:
        x-=1
        if x < 0:
            x = 0
    if action ==2:
        y+=1
        if y >= CELLHEIGHT:
            y = CELLHEIGHT-1
    if action ==3:
        y-=1
        if y < 0:
            y = 0
    return (x,y,state[2],state[3])
```

```python
def reset():
    state0 = states[np.random.choice(len(states))]
    return state0
```

```python
hole_ =[(1,1),(3,2),(0,1),(3,4),(5,3)]
goal_=[(5,0)]
```

### Train Agent befor Statrt Game

```python
n_games = 100000
total_rewards = np.zeros(n_games)
eps_rewards = 0
eps_rewards_cat = 0
alpha =0.15
gamma =0.99
for i in range(n_games):
    
    state = reset()
    # state = get_state(observation[0])
    done = False
    action = maxAction(Q_mouse, state)
    eps_rewards = 0
    for j in range (50):
        state_ = do_action_for_mouse(action,state)
        reward = rewards_mouse(state_, hole_)

        
        # p1 = env.step(action)
        # observation_=p1[0]
        # reward=p1[1]
        # done=p1[2]
        # state_ = get_state(observation_)
        action_ = maxAction(Q_mouse, state_)
        eps_rewards += reward
        Q_mouse[state, action] = Q_mouse[state,action] + alpha*(reward + gamma*Q_mouse[state_,action_] - Q_mouse[state,action])
        state = state_
        action = action_
        # print (reward)
        
        action = maxAction(Q_cat, state)
        state_ = do_action_for_cat(action,state)
        reward = rewards_cat(state_, hole_, goal_)
        action_ = maxAction(Q_cat, state_)
        eps_rewards_cat += reward
        Q_cat[state, action] = Q_cat[state,action] + alpha*(reward + gamma*Q_cat[state_,action_] - Q_cat[state,action])
        state = state_
        action = action_
    total_rewards[i] = eps_rewards
    if i % 10 == 0:
        print('episode ', i, 'score ',total_rewards[i])
```

## Create AND Statrt Game

```python
import random, pygame, sys
from pygame.locals import *
FPS = 2
#             R    G    B
WHITE     = (255, 255, 255)
BLACK     = (  0,   0,   0)
BLACK1     = (  40,   40,   40)
RED       = (155,   20,   20)
GREEN     = (  0, 255,   0)
DARKGREEN = (  0, 155,   0)
DARKGRAY  = ( 40,  40,  40)
BGCOLOR = WHITE

UP = 'up'
DOWN = 'down'
LEFT = 'left'
RIGHT = 'right'


catx_start_point = 100
caty_start_point = 600

catImg = pygame.transform.scale(pygame.image.load("cat1.png"), (100, 100))
catx_ = catx_start_point
caty_ = caty_start_point

mousex_start_point = 400
mousey_start_point = 100

mouseImg = pygame.transform.scale(pygame.image.load("mouse1.jfif"), (100, 100))
mousex_ = mousex_start_point
mousey_ = mousey_start_point


def main():
    global FPSCLOCK, DISPLAYSURF, BASICFONT

    pygame.init()
    FPSCLOCK = pygame.time.Clock()
    DISPLAYSURF = pygame.display.set_mode((WINDOWWIDTH, WINDOWHEIGHT))
    BASICFONT = pygame.font.Font('freesansbold.ttf', 18)
    pygame.display.set_caption('Wormy')
    showStartScreen()
    while True:
        runGame(catx_, caty_, mousex_, mousey_)
        showGameOverScreen

def runGame(catx, caty, mousex, mousey):
    cat_direction = UP
    mouse_direction = RIGHT
    action = 0
    dead_home_mouse = 0
    dead_home_cat = 0
    while True: # main game loop
        for event in pygame.event.get(): # event handling loop
            if event.type == QUIT:
                terminate()
            elif event.type == KEYDOWN:
                if (event.key == K_LEFT):
                    cat_direction = LEFT
                if (event.key == K_RIGHT):
                    cat_direction = RIGHT
                if (event.key == K_UP):
                    cat_direction = UP
                if (event.key == K_DOWN):
                    cat_direction = DOWN
                # if (event.key == K_a):
                #     mouse_direction = LEFT
                # if (event.key == K_d):
                #     mouse_direction = RIGHT
                # if (event.key == K_w):
                #     mouse_direction = UP
                # if (event.key == K_s):
                #     mouse_direction = DOWN
                if event.key == K_ESCAPE:
                    terminate()
        
        if dead_home_cat == 0 :
            if cat_direction == UP:
                caty = caty-100
            elif cat_direction == DOWN:
                caty = caty+100
            elif cat_direction == LEFT:
                catx = catx-100
            elif cat_direction == RIGHT:
                catx = catx+100
        for i in range(len(hole_)):
            if abs(catx-hole_[i][0]*CELLSIZE)<15 and abs(caty-hole_[i][1]*CELLSIZE)<15:
                dead_home_cat = 1

        if catx >= WINDOWWIDTH:
            catx = WINDOWWIDTH-CELLSIZE
        elif catx < 0:
            catx = 0
        elif caty >= WINDOWHEIGHT:
            caty = WINDOWHEIGHT-CELLSIZE
        elif caty < 0:
            caty = 0


        

        state = get_state((catx/CELLSIZE,caty/CELLSIZE,mousex/CELLSIZE,mousey/CELLSIZE))
        action = maxAction(Q_mouse, state)
        reward = rewards_mouse(state_, hole_)
        Q_mouse[state, action] = Q_mouse[state,action] + alpha*(reward + gamma*Q_mouse[state_,action_] - Q_mouse[state,action])
        # print( caty,catx,int(catx/CELLSIZE),int(caty/CELLSIZE))
        # print( mousey,mousex,int(mousex/CELLSIZE),int(mousey/CELLSIZE))
        # print( "****")
        
        
        if dead_home_mouse == 0 :
            if action == 3:
                mousey = mousey-100
            elif action == 2:
                mousey = mousey+100
            elif action == 1:
                mousex = mousex-100
            elif action == 0:
                mousex = mousex+100
        for i in range(len(hole_)):
            if abs(mousex-hole_[i][0]*CELLSIZE)<15 and abs(mousey-hole_[i][1]*CELLSIZE)<15:
                dead_home_mouse = 1

        if mousex >= WINDOWWIDTH:
            mousex = WINDOWWIDTH-CELLSIZE
        elif mousex < 0:
            mousex = 0
        elif mousey >= WINDOWHEIGHT:
            mousey = WINDOWHEIGHT-CELLSIZE
        elif mousey < 0:
            mousey = 0
         
        
        # print( rewards(state, hole_))
        # print( state)
        # print( action)
        # print( "########")





        
        DISPLAYSURF.fill(BGCOLOR)
        drawGrid()
        goal()
        DISPLAYSURF.blit(mouseImg, (mousex, mousey))
        DISPLAYSURF.blit(catImg, (catx, caty))
        hole()
        catch(catx, caty, mousex, mousey)
        dead_home(dead_home_cat)
        win(catx, caty)
        pygame.display.update()
        FPSCLOCK.tick(FPS)
        
def terminate():
    pygame.quit()
    sys.exit()

def catPlace():
    catImg = pygame.transform.scale(pygame.image.load("cat1.png"), (90, 90))
    catx = 10
    caty = 1
def goal():
    pygame. draw. rect(DISPLAYSURF, DARKGREEN, ((goal_[0][0]*CELLSIZE,goal_[0][1]*CELLSIZE),(CELLSIZE,CELLSIZE)))
def win(catx, caty):
    if catx/CELLSIZE==goal_[0][0] and caty/CELLSIZE==goal_[0][1]:
        showWINScreen()
def hole():
    for i in range(len(hole_)):
        pygame. draw. rect(DISPLAYSURF, BLACK1, ((hole_[i][0]*CELLSIZE,hole_[i][1]*CELLSIZE),(CELLSIZE,CELLSIZE)))
    # pygame. draw. rect(SCREEN, RED, ((0,0),(700,50)))

def drawPressKeyMsg():
    pressKeySurf = BASICFONT.render('Press a key to play.', True, DARKGRAY)
    pressKeyRect = pressKeySurf.get_rect()
    pressKeyRect.topleft = (WINDOWWIDTH - 200, WINDOWHEIGHT - 30)
    DISPLAYSURF.blit(pressKeySurf, pressKeyRect)

def checkForKeyPress():
    if len(pygame.event.get(QUIT)) > 0:
        terminate()

    keyUpEvents = pygame.event.get(KEYUP)
    if len(keyUpEvents) == 0:
        return None
    if keyUpEvents[0].key == K_ESCAPE:
        terminate()
    return keyUpEvents[0].key
def showWINScreen():
    gameOverFont = pygame.font.Font('freesansbold.ttf', 150)
    gameSurf = gameOverFont.render('WIN', True, DARKGREEN)
    overSurf = gameOverFont.render(':)', True, DARKGREEN)
    gameRect = gameSurf.get_rect()
    overRect = overSurf.get_rect()
    gameRect.midtop = (WINDOWWIDTH / 2, 10)
    overRect.midtop = (WINDOWWIDTH / 2, gameRect.height + 10 + 25)

    DISPLAYSURF.blit(gameSurf, gameRect)
    DISPLAYSURF.blit(overSurf, overRect)
    drawPressKeyMsg()
    pygame.display.update()
    pygame.time.wait(500)
    checkForKeyPress() # clear out any key presses in the event queue

    while True:
        if checkForKeyPress():
            pygame.event.get() # clear event queue
            return

def showGameOverScreen():
    gameOverFont = pygame.font.Font('freesansbold.ttf', 150)
    gameSurf = gameOverFont.render('Game', True, RED)
    overSurf = gameOverFont.render('Over', True, RED)
    gameRect = gameSurf.get_rect()
    overRect = overSurf.get_rect()
    gameRect.midtop = (WINDOWWIDTH / 2, 10)
    overRect.midtop = (WINDOWWIDTH / 2, gameRect.height + 10 + 25)

    DISPLAYSURF.blit(gameSurf, gameRect)
    DISPLAYSURF.blit(overSurf, overRect)
    drawPressKeyMsg()
    pygame.display.update()
    pygame.time.wait(500)
    checkForKeyPress() # clear out any key presses in the event queue

    while True:
        if checkForKeyPress():
            pygame.event.get() # clear event queue
            return
def dead_home(dead_home_cat):
    if dead_home_cat ==1:
        showGameOverScreen()

def catch(catx, caty, mousex, mousey):
    if abs(catx-mousex)<=1 and abs(caty-mousey)<=1:
        showGameOverScreen()
    
def drawGrid():
    for x in range(0, WINDOWWIDTH, CELLSIZE): # draw vertical lines
        pygame.draw.line(DISPLAYSURF, DARKGRAY, (x, 0), (x, WINDOWHEIGHT))
    for y in range(0, WINDOWHEIGHT, CELLSIZE): # draw horizontal lines
        pygame.draw.line(DISPLAYSURF, DARKGRAY, (0, y), (WINDOWWIDTH, y))


def showStartScreen():
    titleFont = pygame.font.Font('freesansbold.ttf', 100)
    titleSurf1 = titleFont.render('Cat!', True, WHITE, DARKGREEN)
    titleSurf2 = titleFont.render('Mouse!', True, RED)

    degrees1 = 0
    degrees2 = 0
    while True:
        DISPLAYSURF.fill(BGCOLOR)
        rotatedSurf1 = pygame.transform.rotate(titleSurf1, degrees1)
        rotatedRect1 = rotatedSurf1.get_rect()
        rotatedRect1.center = (WINDOWWIDTH / 2, WINDOWHEIGHT / 2-100)
        DISPLAYSURF.blit(rotatedSurf1, rotatedRect1)

        rotatedSurf2 = pygame.transform.rotate(titleSurf2, degrees2)
        rotatedRect2 = rotatedSurf2.get_rect()
        rotatedRect2.center = (WINDOWWIDTH / 2, WINDOWHEIGHT / 2+100)
        DISPLAYSURF.blit(rotatedSurf2, rotatedRect2)

        drawPressKeyMsg()

        if checkForKeyPress():
            pygame.event.get() # clear event queue
            return
        pygame.display.update()
        FPSCLOCK.tick(FPS*15)
        degrees1 += 3 # rotate by 3 degrees each frame
        degrees2 += 7 # rotate by 7 degrees each frame

if __name__ == '__main__':
    main()
```
